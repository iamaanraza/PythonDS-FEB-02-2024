```python
import numpy as np
import plotly.express as px
```

```python
x = np.arange(10)
y = np.random.randint(1, 100, 10)
print(x)
print(y)
px.line( x=x, 
        y=y,
        title = "line graph",
        labels= {'x':'X axis', 'y':'Y axis'})
```

```python
a = np.random.uniform(0, 1, 100)
b = np.random.randn(100)
c = np.random.uniform(0, 1, 100)
print(a.shape, b.shape, c.shape )
px.scatter_3d(
    x=a,
    y=b,
    z=c,
    title="3D Graph"
)
```
